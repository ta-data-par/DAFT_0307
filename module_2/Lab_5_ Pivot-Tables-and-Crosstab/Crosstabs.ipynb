```python
# Import Libraries
import pandas as pd
```

### 1. Import the moovies dataset 

# https://www.kaggle.com/carolzhangdc/imdb-5000-movie-dataset

```python
data = pd.read_csv('movie_metadata.csv')
data.head()
```

### 2. Use crosstabs to show the number of moovies per country

```python
# Use crosstab to show the number of movies per country
movie_counts_per_country = pd.crosstab(index=data['country'], columns='count')

# Display the result
print(movie_counts_per_country)
```

### 3. Create a crosstab to show the number of films per director and genre splitted by country

```python

# Create the crosstab
director_genre_country_counts = pd.crosstab(index=[data['director_name'], data['genres']], columns=data['country'], margins=True, margins_name="Total")

# Display the result
print(director_genre_country_counts)
```

### 4. Average duration of filmes grouped by color

```python
# Group by 'color' and calculate the average duration for each group
average_duration_by_color = data.groupby('color')['duration'].mean()

# Display the result
print(average_duration_by_color)
```

### 5. Average number of likes per director (assuming that total film's likes are counted as a sum of actor's likes)

```python
# Calculate the total likes for each movie by summing up the likes of all actors
data['total_movie_likes'] = data['actor_1_facebook_likes'] + data['actor_2_facebook_likes'] + data['actor_3_facebook_likes']

# Group by 'director_name' and calculate the sum of the total likes for each director
director_total_likes = data.groupby('director_name')['total_movie_likes'].sum()

# Calculate the average number of likes per director
average_likes_per_director = director_total_likes.mean()

# Display the result
print("Average number of likes per director:", average_likes_per_director)
```
